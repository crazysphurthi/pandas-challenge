```python
import pandas as pd
import numpy as np

file_to_load = "purchase_data.csv"

purchase_data = pd.read_csv(file_to_load)
```

```python
total_players = len(purchase_data["SN"].unique())
total_player_df = pd.DataFrame({"Total Players": [total_players]})
total_player_df
```

```python
purchase_data.head()
number_of_unique_items = len(purchase_data["Item ID"].unique())
avg_price = purchase_data["Price"].mean()
number_of_purchases = len(purchase_data["Purchase ID"].unique())
total_revenue = purchase_data["Price"].sum()

purchasing_analysis_df = pd.DataFrame({"Number of Unique Items": [number_of_unique_items], "Average Price": [avg_price], "Number of Purchases": [number_of_purchases], "Total Revenue": [total_revenue]})
purchasing_analysis_df
```

```python
# "Male"
male_players_df = purchase_data.loc[purchase_data["Gender"] == "Male", :]
unique_male_players = male_players_df["SN"].unique()
num_unique_male_players = len(unique_male_players)

# "Female"
female_players_df = purchase_data.loc[purchase_data["Gender"] == "Female", :]
unique_female_players = female_players_df["SN"].unique()
num_unique_female_players = len(unique_female_players)

# "Other / Non-Disclosed"
other_players_df = purchase_data.loc[purchase_data["Gender"] == "Other / Non-Disclosed", :]
unique_other_players = other_players_df["SN"].unique()
num_unique_other_players = len(unique_other_players)
num_unique_other_players

# Percent
total_unique_players = num_unique_male_players + num_unique_female_players + num_unique_other_players
percent_unique_male_players = num_unique_male_players / total_unique_players * 100
percent_unique_female_players = num_unique_female_players / total_unique_players * 100
percent_unique_other_players = num_unique_other_players / total_unique_players * 100

# Summary Table
summary_table = pd.DataFrame({"Gender": ["Male", "Female", "Other / Non-Disclosed"], "Total Count": [num_unique_male_players, num_unique_female_players, num_unique_other_players], "Percentage of Players": [percent_unique_male_players, percent_unique_female_players, percent_unique_other_players]})
summary_table = summary_table.set_index("Gender")
summary_table
```

```python
# "Male"
male_players_df = purchase_data.loc[purchase_data["Gender"] == "Male", :]
male_purchase_count = len(male_players_df["Purchase ID"])
male_avg_purchase_price = male_players_df["Price"].mean()
male_total_purchase_value = male_players_df["Price"].sum()
avg_total_purcase_per_male = male_total_purchase_value/num_unique_male_players

# "Female"
female_players_df = purchase_data.loc[purchase_data["Gender"] == "Female", :]
female_purchase_count = len(female_players_df["Purchase ID"])
female_avg_purchase_price = female_players_df["Price"].mean()
female_total_purchase_value = female_players_df["Price"].sum()
avg_total_purcase_per_female = female_total_purchase_value/num_unique_female_players

# "Male"
other_players_df = purchase_data.loc[purchase_data["Gender"] == "Other / Non-Disclosed", :]
other_purchase_count = len(other_players_df["Purchase ID"])
other_avg_purchase_price = other_players_df["Price"].mean()
other_total_purchase_value = other_players_df["Price"].sum()
avg_total_purcase_per_other = other_total_purchase_value/num_unique_other_players

# Summary Table
summary_table = pd.DataFrame({"Gender": ["Male", "Female", "Other / Non-Disclosed"], 
                             "Purchase Count": [male_purchase_count, female_purchase_count, other_purchase_count],
                             "Average Purchase Price": [male_avg_purchase_price, female_avg_purchase_price, other_avg_purchase_price],
                             "Total Purchase Value": [male_total_purchase_value, female_total_purchase_value, other_total_purchase_value],
                             "Avg Total Purchase per Person": [avg_total_purcase_per_male, avg_total_purcase_per_female, avg_total_purcase_per_other]})
summary_table = summary_table.set_index("Gender")
summary_table
```

```python
# Age Demographics

#print(purchase_data["Age"].min())
#print(purchase_data["Age"].max())

bins = [0, 10, 14, 19, 24, 29, 34, 39, 100]
age_groups = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

purchase_data["Age Group"] = pd.cut(purchase_data["Age"], bins, labels=age_groups)

num_age_10 = len(purchase_data.loc[purchase_data["Age Group"] == "<10", :]["SN"].unique())
num_age_14 = len(purchase_data.loc[purchase_data["Age Group"] == "10-14", :]["SN"].unique())
num_age_19 = len(purchase_data.loc[purchase_data["Age Group"] == "15-19", :]["SN"].unique())
num_age_24 = len(purchase_data.loc[purchase_data["Age Group"] == "20-24", :]["SN"].unique())
num_age_29 = len(purchase_data.loc[purchase_data["Age Group"] == "25-29", :]["SN"].unique())
num_age_34 = len(purchase_data.loc[purchase_data["Age Group"] == "30-34", :]["SN"].unique())
num_age_39 = len(purchase_data.loc[purchase_data["Age Group"] == "35-39", :]["SN"].unique())
num_age_100 = len(purchase_data.loc[purchase_data["Age Group"] == "40+", :]["SN"].unique())

total_unique_people = num_age_10 + num_age_14 + num_age_19 + num_age_24 + num_age_29 + num_age_34 + num_age_39 + num_age_100
pct_age_10 = num_age_10 / total_unique_people * 100
pct_age_14 = num_age_14 / total_unique_people * 100
pct_age_19 = num_age_19 / total_unique_people * 100
pct_age_24 = num_age_24 / total_unique_people * 100
pct_age_29 = num_age_29 / total_unique_people * 100
pct_age_34 = num_age_34 / total_unique_people * 100
pct_age_39 = num_age_39 / total_unique_people * 100
pct_age_100 = num_age_100 / total_unique_people * 100

summary_table = pd.DataFrame({"Age Groups": age_groups,
                             "Total Count": [num_age_10, num_age_14, num_age_19, num_age_24, num_age_29, num_age_34, num_age_39, num_age_100],
                             "Percentage of Players": [pct_age_10, pct_age_14, pct_age_19, pct_age_24, pct_age_29, pct_age_34, pct_age_39, pct_age_100]})
summary_table = summary_table.set_index("Age Groups")
summary_table
```
